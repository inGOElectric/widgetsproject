```python
import mysql.connector
import pandas as pd

# Create a connection to the database
connection = mysql.connector.connect(
    host='database-1.cbjabnlglbz6.ap-south-1.rds.amazonaws.com',
    user='admin',
    password='ingo4321',
    database='ingodata'
)

# Create a cursor
cursor = connection.cursor(dictionary=True)

# Query to get the list of tables
get_tables_query = 'SHOW TABLES'

# Execute the query to get tables
cursor.execute(get_tables_query)
tables = cursor.fetchall()

# Results list to store data
results_list = []

# Iterate through tables and get row counts
for table in tables:
    table_name = list(table.values())[0]
    row_count_query = f"SELECT '{table_name}' AS table_name, COUNT(*) AS row_count FROM ingodata.{table_name} WHERE date_time >= CURDATE() - INTERVAL 30 DAY"
    cursor.execute(row_count_query)
    result = cursor.fetchone()
    results_list.append(result)
cursor.close()
connection.close()
df = pd.DataFrame(results_list)
# Write the DataFrame to an Excel file
df.to_excel(r'C:\Users\Aparna\Downloads\results1.xlsx', index=False)
print('Results have been stored in results1.xlsx')
```
